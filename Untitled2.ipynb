```python
import pandas as pd
df = pd.read_csv('Student performance Dataset.csv')
```

```python
# Check missing values
print(df.isnull().sum())

# Handle missing values using forward fill
df.ffill(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of scores
sns.histplot(df['MathScore'])
plt.show()

sns.boxplot(x='Gender', y='MathScore', data=df)
plt.show()

# Correlation heatmap (select only numeric columns)
numeric_df = df.select_dtypes(include=['number'])  # Only numeric columns
corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd

# Step 1: Convert 'WklyStudyHours' into numeric before encoding
df['WklyStudyHours'] = df['WklyStudyHours'].replace({
    '< 5': 2,
    '05-Oct': 7,
    '> 10': 12
}).infer_objects(copy=False)  # Removes the warning

# Step 2: Convert categorical variables to numeric using one-hot encoding
df_encoded = pd.get_dummies(df, drop_first=True)

# Optional: Verify everything worked
print(df_encoded.head())
print("\nData types after encoding:")
print(df_encoded.dtypes)
```

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop(['MathScore', 'ReadingScore', 'WritingScore'], axis=1)
y_math = df_encoded['MathScore']
y_reading = df_encoded['ReadingScore']
y_writing = df_encoded['WritingScore']

X_train, X_test, y_math_train, y_math_test = train_test_split(X, y_math, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_math_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_math_test, y_pred))
print("R² Score:", r2_score(y_math_test, y_pred))
```

```python
# Step 1: Compute Average Score
df['AverageScore'] = df[['MathScore', 'ReadingScore', 'WritingScore']].mean(axis=1)

# Step 2: Create Grade Class
def grade_class(avg):
    if avg >= 85:
        return 'A'
    elif avg >= 70:
        return 'B'
    else:
        return 'C'

df['FinalGradeClass'] = df['AverageScore'].apply(grade_class)

# Step 3: Encode all relevant data for ML
df['WklyStudyHours'] = df['WklyStudyHours'].replace({
    '< 5': 2,
    '05-Oct': 7,
    '> 10': 12
}).infer_objects(copy=False)

df_encoded = pd.get_dummies(df, drop_first=True)

# Step 4: View updated data
print(df[['MathScore', 'ReadingScore', 'WritingScore', 'AverageScore', 'FinalGradeClass']].head())
```
